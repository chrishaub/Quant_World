```python
import pandas as pd

# create a sample dataframe with dummy variables
df = pd.DataFrame({
    'color': ['blue', 'red', 'blue', 'green', 'red', 'blue', 'green', 'blue'],
    'color_blue': [1, 0, 1, 0, 0, 1, 0, 1],
    'color_red': [0, 1, 0, 0, 1, 0, 0, 0],
    'color_green': [0, 0, 0, 1, 0, 0, 1, 0],
    'color_yellow': [0, 0, 0, 0, 0, 0, 0, 0]
})



df
```

```python
# get the columns with the same values
duplicate_columns = set()
for i in range(df.shape[1]):
    col_i = df.iloc[:, i]
    for j in range(i+1, df.shape[1]):
        col_j = df.iloc[:, j]
        if col_i.equals(col_j):
            duplicate_columns.add(df.columns[j])
```

```python
# combine the duplicate columns
df = df.drop(duplicate_columns, axis=1)

# test the result
df
```
